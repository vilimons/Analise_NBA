## Web Scraping e Configuração do Ambiente

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html#per_game_stats'
html = pd.read_html(url, header=0)
df2022 = html[0]

df2022
```

## Análise exploratória de dados (EDA)

Leitura de dados.

```python
df = pd.read_csv('nba2022.csv')
```

Removendo linhas inúteis.

```python
df = df.drop(df.index[[22, 43, 64, 89, 110, 131, 152, 178, 205, 226, 247, 272, 296, 321, 346, 371, 392, 413, 436, 457, 478, 499, 520, 546, 569, 592, 617, 642, 663]])
```

Removeremos a coluna 'RK' para evitar confusão com o index.

```python
df = df.drop(['Rk'], axis=1)
```

Visualizando os tipos de dados de cada coluna.

```python
df.dtypes
```

#### Como observado acima, todas as colunas são do tipo 'Object'.

- Para poder realizar a análise, precisamos de colunas do tipo numérico.

- Separei e converti para numérico as colunas que estavam com o tipo errado.

```python
cols =  ["Age", "G", "GS", "MP", "FG", "FGA", "FG%", "3P", "3PA", "3P%", "2P", "2PA", "2P%", "eFG%", "FT", "FTA", "FT%", "ORB", "DRB", "TRB", "AST", "STL", "BLK", "TOV", "PF", "PTS"]

df[cols] = df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

Checando se a conversão foi bem sucedida.

```python
df.dtypes
```

Mudando uma opção do pandas para mostrar todas as colunas.

```python
pd.set_option('display.max_columns', None)
```

## Análise

### Seleção Condicional

Para realizar uma análise de dados exploratória, é importante separar o dataframe em subconjuntos de dados para fazermos a análise ou comparações.

**Qual jogador marcou o maior número de pontos por jogo?**

```python
df[df.PTS == df.PTS.max()]
```

**De qual time é o maior pontuador?**

```python
MaiorPontuador = df[df.PTS == df.PTS.max()]
MaiorPontuador.Tm
```

 O jogador é do time Brooklyn Nets

$\;\;\;\;\;\;$

**De qual posição é o maior pontuador?**

```python
MaiorPontuador.Pos
```

O maior pontuador é um Ala-Pivô (ou Power Forward)

$\;\;\;\;\;\;$


**Quantos jogos o melhor pontuador jogou?**

```python
MaiorPontuador.G
```

#### Jogadores que marcaram mais de 20 pontos por partida.

```python
df[df.PTS > 20]
```

**Jogador que mais marcou no arremesso de 3 pontos.**

```python
df[df['3P'] == df['3P'].max()]
```

**Jogador que deu o maior número de assistências.**

```python
df[df['AST'] == df['AST'].max()]
```

### Agrupando jogadores do time Phoenix Suns.

```python
PHO = df.groupby('Tm').get_group('PHO')
```

**Jogador que mais pontuou pelo Phoenix Suns.**

```python
PHO[PHO.PTS == PHO.PTS.max()]
```

**Jogador com maior número de arremesso de 3 pontos pelo Phoenix Suns.**

```python
PHO[PHO['3P'] == PHO['3P'].max()]
```

**Jogador que mais deu assistências pelo Phoenix Suns.**

```python
PHO[PHO['AST'] == PHO['AST'].max()]
```

### Analisando a performance de cada posição.

**Qual posição marcou mais pontos?**

```python
df.groupby('Pos').PTS.describe()
```

### Agora vamos fazer análises apenas das 5 posições tradicionais.

Todos os jogadores das 5 posições tradicionais.

```python
positions = ['C', 'PF', 'SF', 'PG', 'SG']
POS = df[df['Pos'].isin(positions)]
POS
```

Entre as 5 posições tradicionais, qual marcou mais pontos?

```python
POS.groupby('Pos').PTS.describe()
```

## Gráficos

#### Vamos fazer a mesma análise, mas mostrando as respostas através de gráficos.

Para facilitar, separei as 5 posições tradicionais em um subconjunto de dados.

```python
PTS = df[['Pos', 'PTS']]

positions = ['C', 'PF', 'SF', 'PG', 'SG']
PTS = PTS[PTS['Pos'].isin(positions)]

PTS
```

```python
g = sns.FacetGrid(PTS, col="Pos")
g.map(plt.hist, "PTS")
```

```python
PTS.boxplot(column='PTS', by='Pos')
```

```python
sns.boxplot(x = 'Pos', y = 'PTS', data = PTS) 
```

```python
sns.boxplot(x = 'Pos', y = 'PTS', data = PTS) 
sns.stripplot(x = 'Pos', y = 'PTS', data = PTS,
              jitter=True, 
              marker='o',
              alpha=0.8, 
              color="black")
```
